#THE BASICS :

#initialize array

# with assigning to dtype i,e keyword argument we  can make it possible to store our data in specified number of bytes
#but it need to fit into that specified byte of storage;

```python

a=np.array([1,2,3],dtype="int16")  
print("one dimensional array :",a)

b=np.array([[1.0,2.0,3.0,5.8],[2.1,3,4.2,4.5]])
print("two dimensional array :",b)
```

#Get Dimension

```python

print("dimension of a :",a.ndim)
print("dimension of b :",b.ndim)
```

#Get Shape

#its a vector so its tellings us one dimension 

```python

print("shape of a :",a.shape) 
print("shape of b :",b.shape)
```

#Get  Type

```python

print("a type is :",a.dtype)
```

#Get Size

```python

print("size of a :",a.size) # # of element in the array.i,e length of array;
```

#total size
# size of single element in the array .i,e in bytes;

```python

print("a item size :",a.itemsize) 
print("total size is a:",a.size*a.itemsize)
```
